```python
import polychrom_hoomd.render as render
from lem import (
    config,
    populate_angles_to_snapshot,
    populate_bonds_to_snapshot,
    populate_particles_to_snapshot,
    set_snapshot_box,
)

snapshot = set_snapshot_box(config.cpcdh())
snapshot = populate_particles_to_snapshot(config.cpcdh(), snapshot)
snapshot = populate_bonds_to_snapshot(config.cpcdh(), snapshot)
snapshot = populate_angles_to_snapshot(config.cpcdh(), snapshot)

# Visualize starting conformation using the Fresnel backend (A compartments in blue, B in red)
render.fresnel(snapshot, show="compartments", cmap="coolwarm").static(pathtrace=True)
```
